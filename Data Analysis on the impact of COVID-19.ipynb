### Importing important libraries

```python
## Loading the data in dataframe with the help of reading csv file
df = pd.read_csv('owid-covid-data.csv')
df.head()
```

```python
## Determining the total number of deaths due to COVID-19 all around the world
total_COVID_deaths = np.sum(df['new_deaths'])
total_COVID_deaths
```

```python
## Determining the Data collection time period
a = np.min(df['date'])
b = np.max(df['date'])
print('The data available is from {} to {} time period.'.format(a,b))
```

### Data Cleaning
- In this step we will extract only useful data and remove all the unnecessary data from the dataframe.

```python
## Fetching the total number of columns from the data and storing it into the list
total_columns = list(df.columns)   
total_columns
```

```python
### Now we will create a new dataframe consisting the columns only that are important and useful
useful_data = df[['location','date','new_cases','new_deaths','icu_patients','hosp_patients','total_tests','new_tests',
                 'positive_rate','total_vaccinations','people_vaccinated','people_fully_vaccinated','stringency_index',
                 'aged_65_older','aged_70_older','cardiovasc_death_rate','diabetes_prevalence','female_smokers','male_smokers',
                'population']]
```

```python
## Adding one column in our dataframe
useful_data['life_expectancy'] = df.life_expectancy
useful_data.head()
```

```python
## These are the sorted columns out of the total_columns in 'owid_covid_data'. 
## Only these columns are useful so we will use only these columns to analyze the data. 
useful_columns = useful_data.columns
useful_columns
```

```python
## Checking number of NULL values in our useful dataset
useful_data.isnull().sum()
```

##### - There are so many NULL values in our dataset, so before analyzing the data we have to fill these null values with the values appropriate to particular columns.

```python
useful_data.head()
```

##### A). Filling the NULL values in the columns stated below with '0' 
1. new_cases
2. new_deaths
3. icu_patients
4. hosp_patients
5. total_tests
6. new_tests
7. total_vaccinations
8. people_vaccinated
9. people_fully_vaccinated
10. female_smokers
11. male_smokers

```python
useful_data[['new_cases','new_deaths','icu_patients','hosp_patients','total_tests','new_tests','total_vaccinations',
             'people_vaccinated','people_fully_vaccinated','female_smokers','male_smokers']] = useful_data[['new_cases',
            'new_deaths','icu_patients','hosp_patients','total_tests','new_tests','total_vaccinations',
             'people_vaccinated','people_fully_vaccinated','female_smokers','male_smokers']].fillna(value=0)
```

```python
## Now again checking the null values. 
## We can see that the number of NULL values in above stated columns are 0
useful_data.isnull().sum()
```

##### B). Filling the NULL values in the columns stated below with 'Median'
1. positive_rate
2. stringency_index
3. aged_65_older
4. aged_70_older

```python
## First of all we will obtain the median values of all the columns
w = useful_data['positive_rate'].median()
x = useful_data['stringency_index'].median()
y = useful_data['aged_65_older'].median()
z = useful_data['aged_70_older'].median()
```

```python
## Now we will create a dictionary with keys = column names and values = median values of respective columns
median_values = {'positive_rate':w, 'stringency_index':x,'aged_65_older':y,'aged_70_older':z}
```

```python
## Now we will use fillna function and insert the dictionary as value
useful_data[['positive_rate','stringency_index','aged_65_older','aged_70_older']] = useful_data[['positive_rate',
          'stringency_index','aged_65_older','aged_70_older']].fillna(median_values)
```

##### C). Filling the NULL values in the columns stated below with 'Mean'
1. cardiovasc_death_rate
2. diabetes_prevalence
3. life_expectancy

```python
## First of all we will obtain the mean values of all the columns
c = useful_data['cardiovasc_death_rate'].mean()
d = useful_data['diabetes_prevalence'].mean()
e = useful_data['life_expectancy'].mean()
```

```python
## Now we will create a dictionary with keys = column names and values = mean values of respective columns
mean_values = {'cardiovasc_death_rate':c,'diabetes_prevalence':d,'life_expectancy':e}
```

```python
## Now we will use fillna function and insert the dictionary as value
useful_data[['cardiovasc_death_rate','diabetes_prevalence','life_expectancy']] = useful_data[['cardiovasc_death_rate',
            'diabetes_prevalence','life_expectancy']].fillna(mean_values)
```

```python
useful_data.isnull().sum()
```

#### Now we can see that there are only 1071 null values in 'population' column which is very small in comparison to whole data.
#### So we will drop all the remaining null values because it is not much important and it will not create any disruption in our dataframe.

```python
np.shape(useful_data)   ## No. of rows before dropping null values
```

```python
useful_data.dropna(inplace = True) ## No. of rows after dropping null values
np.shape(useful_data)
```

##### Finally, now our dataframe is ready to be analyzed

```python
## Exporting our data to csv file
useful_data.to_csv('useful_owid_data.csv', index = False)
```

### Data Analyzation
- In this step we will analyze our useful data and create some useful dataframes and plot different graphs

```python
## Grouping the dataframe by locations for fetching total number of cases
total_cases = useful_data.groupby('location')['new_cases'].sum()
total_cases = total_cases.sort_values(ascending=False)   ## sorting from highest to lowest 

## Picking randomly 5 countries
most_affected_countries = total_cases[10:15].index
most_affected_countries
```

```python
cases = total_cases[10:15].values # Fetching number of Cases in India, France, Germany, Brazil, Japan 
print(cases)
```

```python
## Grouping the dataframe by locations for fetching total number of deaths
total_deaths = useful_data.groupby('location')['new_deaths'].sum()
deaths = total_deaths[most_affected_countries].values # Number of Deaths in India, France, Germany, Brazil, Japan 
print(deaths)
```

```python
## Creating a dataframe consisting 5 countries and number of cases and deaths in that counrties
cases_deaths_df = pd.DataFrame({'Country': most_affected_countries, 'Total Cases': cases, 'Total Deaths': deaths})
cases_deaths_df
```

```python
## Plotting vertical bar chart of Country vs. Total Cases
sns.barplot(x = 'Country', y = 'Total Cases', data=cases_deaths_df, palette = 'crest')

plt.title('Total number of Cases')
```

##### From above chart we can see that out of these 5 countries, India has the highest number of cases.

```python
## Plotting vertical bar chart of Country vs. Total Deaths
sns.barplot(x = 'Country', y = 'Total Deaths', data=cases_deaths_df, palette = 'flare')
    
plt.title('Total number of Deaths')
```

##### From above chart we can see that out of these 5 countries, Brazil has the highest number of Deaths.

#### Now as we have found out the relation of total cases and deaths between some countries, we will move ahead and do further analysis

```python
## Finding the testing rate of the some countries
total_tests = df.groupby('location')['new_tests'].sum().values
population = df.groupby('location')['population'].nth(-1)   ## nth(-1) = the last value of population of particular country
testing_rate = (total_tests/population).sort_values(ascending = False)[:10]  ## highest to lowest
```

```python
## Creating a dataframe consisting columns - Country and testting_rate
testing_rate_df = pd.DataFrame({'Country':testing_rate.index,'testing_rate(total_tests/population)':testing_rate.values})
testing_rate_df
```

```python
## Plotting a horizontal bar plot
sns.barplot(x = testing_rate.values, y = testing_rate.index, palette="vlag")
plt.title('Testing Rate (Top 10 Countries)')
plt.xlabel('Total Tests / Total Population')
plt.ylabel('Country')
```

##### Cyprus is the country with the highest testing rate among all the countries.

```python
## Finding 5 countries and thier number of vaccination doses administered
vaccinations = useful_data.groupby('location')['total_vaccinations'].nth(-1)
vaccinations = vaccinations.sort_values(ascending=False)
doses_administered_countries = vaccinations[10:15].index ## Countries with doses administered

## Using for loop to print all the countries
for countries in doses_administered_countries:
    print(countries)
```

```python
vaccination_doses_administered = vaccinations[doses_administered_countries].values  ## Number of doses administered

## Using for loop to print number of administered doses from all the countries
for doses in vaccination_doses_administered:
    print(doses)
```

```python
## Creating a dataframe for regions/countries and the vaccination doses administered
vaccination_doses_administered_df = pd.DataFrame({'Country':doses_administered_countries, 'vaccination_doses_administered':vaccination_doses_administered}) 
vaccination_doses_administered_df
```

```python
## Plotting the values of above stated dataframe in the pie chart
labels = vaccinations[doses_administered_countries].index
x = vaccinations[doses_administered_countries].values
plt.pie(x, labels = labels, autopct='%.1f')
plt.title('Number of vaccination doses administered')
```
